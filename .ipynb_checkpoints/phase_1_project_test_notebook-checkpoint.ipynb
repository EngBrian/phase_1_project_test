# 3. Perfom Data Cleaning on the aviaton_data DataFrame

```python
#Drop all the duplicates in the Dataframe
aviation_data.drop_duplicates()
```

```python
#Check for the columns with missing values and their sums

aviation_data.isna().sum()
```

```python
# # Drop columns with more than 50% missing values
aviation_data = aviation_data.drop(columns = ['Latitude', 'Longitude', 'Aircraft.Category', 'FAR.Description', 'Schedule','Air.carrier' ])
aviation_data
```

```python
#Dropping Rows with Missing Values less than 90% in Specific Columns
aviation_data = aviation_data.dropna(subset = ['Location', 'Country', 'Injury.Severity', 'Aircraft.damage',
                                               'Registration.Number','Amateur.Built', 'Make', 'Model', 'Engine.Type','Purpose.of.flight',
                                              'Total.Uninjured', 'Weather.Condition', 'Report.Status'])
aviation_data
```

```python
aviation_data.isna().sum()
```

```python
#Replacing Missing Values with a Placeholder
aviation_data[['Airport.Code', 'Airport.Name', 'Publication.Date']] = aviation_data[['Airport.Code', 'Airport.Name', 'Publication.Date']].fillna('NaN')
# aviation_data[['Airport.Code', 'Airport.Name']]
```

```python
aviation_data[['Number.of.Engines','Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries']] = aviation_data[['Number.of.Engines','Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries']].fillna('mean')
```

```python
aviation_data['Broad.phase.of.flight'] = aviation_data['Broad.phase.of.flight'].fillna('mode')
```

```python
aviation_data.isna().sum()
```
